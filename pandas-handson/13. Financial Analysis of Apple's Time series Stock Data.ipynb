```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas_datareader.data as web
import datetime

stocks = ['AAPL', 'MSFT']
start_date = datetime.datetime(2013,1,1)
end_date = datetime.datetime(2015,12,31)
google_finace = web.DataReader(stocks, 'google', start_date, end_date)
google_finace1 = [web.DataReader(ticker, 'google', start_date, end_date) for ticker in stocks]
# google_finace1
final_data = pd.concat(google_finace1, keys=stocks, names=['Ticker'])
final_data
```

### Let's plot closing prices

```python
apple_closing_prices = final_data[['Close']]
apple_closing_prices.head()
```

### We need to treats date as a column for our plot. So, let's reset the index


```python
data = apple_closing_prices.reset_index()
data.head()
```

```python
# data.set_index('Date', inplace=True)
data[data['Ticker'] == 'AAPL'].plot(figsize=(10,8))
```

### Since, we have to see the closing prices of individual firms, let's pivot 'Date' into the index and each Ticker value into columns

```python
daily_close_prices = data.pivot('Date', 'Ticker','Close')
```

```python
daily_close_prices.head()
```

### Now, our data is ready !!!
#### We can plot the closing prices for Apple

```python
daily_close_prices['AAPL'].plot(figsize=(15,9))
```

### What about comparing closing prices for both firms ?

1. We can see that the fluctuations on stock closing prices in Apple is high. Whereas, the price is not much fluctuated for Microsoft

```python
daily_close_prices.plot(figsize=(15,9))
```
